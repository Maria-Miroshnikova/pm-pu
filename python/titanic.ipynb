```python
data['Pclass'] = data['Pclass'].replace(to_replace=[1, 2, 3], value=['элита', 'работяги', 'средний'])
#print(data)

data = data.assign(Expensive = list([0] * data.shape[0]))
data.loc[data.Fare > 25] = data.loc[data.Fare > 25].replace(to_replace=0, value=1)

print(data)
```
